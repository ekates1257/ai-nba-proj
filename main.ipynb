```python
# Gather player data and store in list
from nba_api.stats.endpoints import playercareerstats, teamyearbyyearstats
from nba_api.stats.static import players
import pandas as pd
import time

def get_win_p(team_id, season):
    try:
        team_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=team_id)
        df = team_stats.get_data_frames()[0]
        df = df[df['LEAGUE_ID'] == '00']
        season_row = df[df['YEAR'] == season]
        if season_row.empty:
            return None
        wins = season_row.iloc[0]['WINS']
        losses = season_row.iloc[0]['LOSSES']
        return wins / (wins + losses)
    except:
        return None

all_players = players.get_active_players()
player_data = []

print("Gathering player data for...")
for p in all_players:
    try:
        player_id = p['id']
        full_name = p['full_name']
        stats = playercareerstats.PlayerCareerStats(player_id=player_id)

        # take only players with 2 years in nba
        df = stats.get_data_frames()[0]
        df = df[df['LEAGUE_ID'] == '00']
        if len(df) < 2:
            continue
        
        # gather data from rookie and sophomore seasons
        r_season = df.iloc[0]
        s_season = df.iloc[1]

        # rookie_win_p = get_win_p(r_season['TEAM_ID'], r_season['SEASON_ID'][:4])
        # soph_win_p = get_win_p(s_season['TEAM_ID'], s_season['SEASON_ID'][:4])
        # if rookie_win_p is None or soph_win_p is None:
        #     print(f"Win P for {full_name}")
        #     continue

        gp_r = r_season['GP']
        gp_s = s_season['GP']
        # skip players who didn't play in one of their first two years
        if gp_r == 0 or gp_s == 0:
            print(f"Games played for {full_name}")
            continue

        data = {
            'Name': full_name,
            'ID': player_id,
            'PTS_r': r_season['PTS'] / gp_r,
            'AST_r': r_season['AST'] / gp_r,
            'REB_r': r_season['REB'] / gp_r,
            'TOV_r': r_season['TOV'] / gp_r,
            'FG3M_r': r_season['FG3M'] / gp_r,
            'MIN_r': r_season['MIN'] / gp_r,
            # 'WIN%_r': rookie_win_p,
            'PTS_s': s_season['PTS'] / gp_s,
            'AST_s': s_season['AST'] / gp_s,
            'REB_s': s_season['REB'] / gp_s
            # 'WIN%_s': soph_win_p
        }

        player_data.append(data)

        print(f"Success for player: {p['full_name']}")
        time.sleep(0.6)

    except Exception as e:
        print(f"Failed for player {p['full_name']}: {e}")
```

```python
# Turn player_data into pandas dataframe
df = pd.DataFrame(player_data)
df.dropna(inplace=True)
```
